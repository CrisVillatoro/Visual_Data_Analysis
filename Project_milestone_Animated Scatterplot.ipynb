# Visual Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import imageio.v2 as imageio

```

```python
#Data/gapminder_total_fertility.csv into pandas.


fert = pd.read_csv('gapminder_total_fertility.csv', index_col=0)
fert
```

```python
# Save it in a DataFrame

life = pd.read_excel('gapminder_lifeexpectancy.xlsx', index_col=0)
life
```

```python
#Check the shape of both tables:

print(f'The shape of the life expectacy table is {life.shape}')
print(f'The shape of the fertility table is {fert.shape}')
```

```python
#Columns
fert.columns
```

```python
#Columns
life.columns
```

```python
#One table has columns as strings, the other has integer numbers. 
#To merge the tables, we need to convert both into the same format:

fert.columns = fert.columns.astype(int)
```

```python
#Check the transformation
fert.columns
```

```python
#Check the index
fert.index
```

```python
# Set the new index

fert.index.name = 'country'
```

```python
# Converting the table into a long format

fert = fert.reset_index()
```

```python
# Melting the features

fert = fert.melt(id_vars='country', var_name='year', value_name='Total fertility rate')
```

```python
#Same procedure for the life expectancy data 
life.index
```

```python
life.index.name = 'country'
```

```python
life = life.reset_index()
```

```python
life = life.melt(id_vars='country', var_name='year', value_name='Life expectancy')
```

```python
#Population data from the file data/gapminder_population.xlsx.
population = pd.read_excel('gapminder_population.xlsx', index_col=0)
population
```

```python
print(population.shape)
population.columns
```

```python

population.columns = population.columns.astype(int)
population.index.name = 'country'
population = population.reset_index()
population = population.melt(id_vars='country', var_name='year', value_name='Total population')
```

```python
continents = pd.read_csv('continents.csv', sep=';')
continents
```

```python

df2 = fert.merge(population)
df1 = df2.merge(life)
df = df1.merge(continents, on='country')
df
df.sort_values(by=['year'])
df.iloc[:, [0,5,1,2,3,4]]
```

```python
df = df.rename(columns={'continent':'Continent'})
df
df.iloc[:, [0,5,1,2,3,4]]
```

```python
#Selecting specific columns and plotting 

df_subset = df.loc[df['year'] == 1960]
sns.set_context("paper")

sns.scatterplot(x='Life expectancy', y='Total fertility rate', 
                    data = df_subset, alpha=0.6, size ='Total population', hue='Continent',legend=False, 
                palette='bright', sizes=(150, 1500), linewidth=1)

plt.axis((10.0, 90.0, 0.0, 10.0))

plt.title(f'Correlation between life expectancy and fertility');
```

```python
df_subset = df.loc[df['year'] == 2014]
sns.set_theme()
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
sns.color_palette("Set2")

f, ax = plt.subplots(figsize=(7.5, 7.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = df['Continent']

sns.scatterplot(ax=ax, x='Life expectancy', y='Total fertility rate',
            data=df_subset, size=df.loc[df['year'] == 2014]['Total population'], alpha=1, hue='Continent', hue_order=clarity_ranking, palette='bright', 
                sizes=(350, 3500), linewidth=1, legend = 'brief')
h,l = ax.get_legend_handles_labels()
plt.legend(h[0:7], l[0:7], bbox_to_anchor=(1.5,1), loc=1, borderaxespad=0.2, fontsize =13)
plt.title(f'Correlation between life expectancy and fertility')

ax.text(40.8, 0.9, "Year 2014", horizontalalignment='left', size='xx-large', color='black', weight='semibold')
plt.axis((30.0, 90.0, 0.0, 10.0));

```

```python
#One plot for each year between 1960 and 2015 (before that, the data contains too many gaps).


df_year_subset= df.loc[(df['year'] >= 1960) & (df['year'] <= 2015)]

for year in df_year_subset['year'].unique():
    f, ax = plt.subplots(figsize=(7.5, 7.5))
    sns.set_theme()
    sns.set_style("darkgrid", {"axes.facecolor": ".9"})
    sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
    sns.color_palette("Set2")
    
    sns.scatterplot(ax=ax, x='Life expectancy', y='Total fertility rate',
            data=df_year_subset[df_year_subset['year'] == year], 
                    size=df_year_subset[df_year_subset['year'] == year]['Total population'], 
                    alpha=1, hue='Continent', hue_order=clarity_ranking, palette='bright', 
                sizes=(350, 3500), linewidth=1, legend = 'brief')
    
    h,l = ax.get_legend_handles_labels()
    plt.legend(h[0:7], l[0:7], bbox_to_anchor=(1.1,1), loc=1, borderaxespad=0, fontsize =13)
    plt.title(f'Correlation between life expectancy and fertility {year}')
    plt.axis((20.0, 90.0, 0.0, 10.0))
    
    ax.text(40.8, 0.9, f"Year {year}", horizontalalignment='left', size='xx-large', color='black', weight='semibold')
    
    plt.savefig(f'Life_exp_{year}.png') 
    plt.close('all');
 
```

```python
#Generate the GIF


images = []

for year in range(1960, 2015):
    filename = 'Life_exp_{}.png'.format(year)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=5)
```
