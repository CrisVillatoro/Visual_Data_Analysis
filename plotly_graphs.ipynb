```python
import pandas as pd

fert = pd.read_csv('gapminder_total_fertility.csv', index_col=0)
fert
```

```python
life = pd.read_excel('gapminder_lifeexpectancy.xlsx', index_col=0)
life
```

```python
print(f'The shape of the life expectacy table is {life.shape}')
print(f'The shape of the fertility table is {fert.shape}')
```

```python
fert.columns
```

```python
life.columns
```

```python
fert.columns = fert.columns.astype(int)
```

```python
fert.index.name = 'country'
fert = fert.reset_index()
fert = fert.melt(id_vars='country', var_name='year', value_name='Total fertility rate')
```

```python
life.index.name = 'country'
life = life.reset_index()
life = life.melt(id_vars='country', var_name='year', value_name='Life expectancy')
```

```python
population = pd.read_excel('gapminder_population.xlsx', index_col=0)
population
```

```python
population.columns = population.columns.astype(int)
population.index.name = 'country'
population = population.reset_index()
population = population.melt(id_vars='country', var_name='year', value_name='Total population')
```

```python
continents = pd.read_csv('continents.csv', sep=';')
continents
```

### Merging the tables 

```python
df2 = fert.merge(population)
df1 = df2.merge(life)
df = df1.merge(continents, on='country')


df.sort_values(by=['year'])
df = df.rename(columns={'country':'Country', 'year': 'Year', 'continent':'Continent'})
df

df.iloc[:, [0,5,1,2,3,4]]
```

```python
iso_alpha= pd.read_csv('countries_continents_codes_flags_url.csv')
iso_alpha = iso_alpha.rename(columns={'country':'Country'})
iso_alpha
```

```python
df_location = df.merge(iso_alpha, on='Country')
df_location
```

```python
import plotly.express as px
```

```python
average_score = df.groupby(['Country'])['Life expectancy'].mean().sort_values(ascending=False)
average_score_df = pd.DataFrame({'Country':average_score.index, 'Average life expectancy':average_score.values})
```

```python
geo_df = df_location
geo_df = geo_df[['Country', 'Continent', 'alpha-3']].drop_duplicates()
merged_df = average_score_df.merge(geo_df, left_on='Country', right_on='Country')
```

```python
import plotly.express as px

fig = px.scatter_geo(df_location, locations="alpha-3",
                     color="Continent", # which column to use to set the color of markers
                     hover_name="Country", # column added to hover information
                     size="Total population", # size of markers
                     projection="natural earth",  width=1000,
                    height=500)
fig.write_html("3d_plot.html")
fig.show()
```

```python
import plotly.express as px

fig = px.choropleth(df_location, locations="alpha-3",
                    color="Life expectancy",
                    hover_name="Country", # column to add to hover information
                    color_continuous_scale=px.colors.sequential.Plasma,  width=1000,
                    height=500, title = 'Life expectancy in 2015')

fig.show()
```
